```python
import numpy as np
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 3, figsize=(15, 10), tight_layout=True)

y0 = [1, 2, -5, 2]
axes[0, 0].plot(y0)
axes[0, 0].set_title('Line Graph')

x1 = np.linspace(0, 10, 100)
y1 = 2 + 2 * np.sin(2 * x1)
axes[0, 1].plot(x1, y1)
axes[0, 1].set_title('Sin Function')

x2 = np.random.normal(15, 5, 2000)
axes[0,2].hist(x2, bins=30, alpha=0.7)
axes[0, 2].set_title('Histogram')

x3 = 4 + np.random.normal(0, 2, 60)
y3 = 4 + np.random.normal(0, 2, len(x3))
axes[1, 0].scatter(x3, y3, alpha=0.7)
axes[1, 0].set_title('Scatter Plot')

plt.show()
```
